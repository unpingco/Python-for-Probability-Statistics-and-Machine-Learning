```python
class User:
  def __init__(self, user):
     self.user = user
     self.id = user["id"]
     self.interests = []
     self.name = user["name"]
     try:
       self.friends = user["friends"]
     except:
      self.friends = []
  def add(self, user):
    if type(self.user) != type(user):
      return
    self.friends.append(user)

  def isFriendofFriend(self, user_id):
    return user_id in self.getFriends()

  def getFriends(self):
    friends_ids = []
    for i in self.friends:
      friends_ids.append(i.id)

    return friends_ids
  
  def getNumofFriends(self):
    return len(self.getFriends())

  def pop(self, id):
    for user in self.friends:
      if user.id == id:
        self.friends.remove(user)
        return

users = []
for user in undefinedusers:
  users.append(User(user))

for i,j in friendships:
  for user in users:
    if user.id == i:
      user_i = user
    if user.id == j:
      user_j = user
  if user_i == user_j:
    continue
  user_i.friends.append(user_j)
  user_j.friends.append(user_i)

```

```python
from __future__ import division
# Total connections
total_connections = sum(user.getNumofFriends() for user in users)
print(total_connections)

 # integer division is lame
num_users = len(users) # length of the users list
avg_connections = total_connections / num_users
print(avg_connections)

# Sort by number of friends table
users_byNumofFriends = [(user.id, user.getNumofFriends()) for user in users]
# print(users_byNumofFriends[0])
sorted(users_byNumofFriends, key=lambda row: row[1], reverse=True
```
